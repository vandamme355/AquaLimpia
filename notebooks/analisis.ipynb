# Análisis de Plantas - AquaLimpia S.A.

```
import pandas as pd
import matplotlib.pyplot as plt
```

## Carga de datos

```
df = pd.read_excel("dataset_set_A_aguas_residuales.xlsx")
df.head()
```

## Exploración inicial

```
df.info()
df.isnull().sum()
```

## Limpieza de datos

```
df["dbo_entrada"] = df["dbo_entrada"].fillna(df["dbo_entrada"].mean())
df["dbo_salida"] = df["dbo_salida"].fillna(df["dbo_salida"].mean())
```

## Cálculo de eficiencia

```
df["eficiencia"] = ((df["dbo_entrada"] - df["dbo_salida"]) / df["dbo_entrada"]) * 100
df.head()
```

## Análisis por planta

```
resumen = df.groupby("planta")[["eficiencia", "dbo_salida"]].mean()
resumen
```

## Visualización

```
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.bar(resumen.index, resumen["eficiencia"])
plt.title("Eficiencia promedio por planta")

plt.subplot(1,2,2)
plt.bar(resumen.index, resumen["dbo_salida"])
plt.title("DBO salida promedio por planta")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusión
El análisis permite identificar diferencias en la eficiencia entre plantas y posibles riesgos de incumplimiento.
